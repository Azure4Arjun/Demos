# What are Jupyter Notebooks?

Basically, they are just a JSON file that can contain a mix of two kinds cells:

-   text cells  
    The text cells hold markdown code.  
    (You're reading from a text cell right now.)

-   code cells  
    The code cells can hold code for a large variety of different languages.



As of right now, Jupyter Notebooks are limited to running a single kernel at a time.  However you have the option of calling code in a different language, via a "Magic".  More on that later.

## Storing results

One of the key features of Jupyter Notebooks is that the can store the results of the code you run, in the Notebook.  If you download this Notebook, you'll see the same results I see right now, until you either run the code cell yourself, or clear the contents.
